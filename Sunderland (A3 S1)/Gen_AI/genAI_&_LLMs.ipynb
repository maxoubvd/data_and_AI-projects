# Qwen 2.5B

```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
model_name = "Qwen/Qwen2.5-0.5B-Instruct"
model = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype="auto", device_map="auto")
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
# Prompt setup
prompt = "Give me a short introduction to large language models (less than 30 words)."
messages = [
  {"role": "system", "content": "You are Qwen, created by Alibaba Cloud. You are a helpful assistant."},
  {"role": "user", "content": prompt}
]
```

```python
# Prepare input text
input_text = tokenizer.apply_chat_template(messages, tokenize=False,
add_generation_prompt=True)
model_inputs = tokenizer([input_text], return_tensors="pt").to(model.device)
```

```python
# Generation parameters
generation_params = {"temperature": 0.7, "top_k": 50, "top_p": 0.9, "max_new_tokens": 150}
```

```python
# Generate response
generated_ids = model.generate(**model_inputs, **generation_params)
```

```python
# Decode and display response
response = tokenizer.decode(generated_ids[0], skip_special_tokens=True)
print(f"Response:\n{response}")
```

# Falcon 7B

```python
model_name = "tiiuae/falcon-7b"
model = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype="auto", device_map="auto")
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
prompt = "Give me a short introduction to large language models (less than 30 words)."
messages = [
    {"role": "system", "content": "You are a helpful assistant."},
    {"role": "user", "content": prompt}
]
```

```python
# Prepare input text manually
input_text = "System: You are a helpful assistant.\nUser: Give me a short introduction to large language models (less than 30 words).\nAssistant:"
model_inputs = tokenizer([input_text], return_tensors="pt").to(model.device)
```

```python
# Generation parameters
generation_params = {"temperature": 0.7, "top_k": 50, "top_p": 0.9, "max_new_tokens": 150}
# Generate response
generated_ids = model.generate(**model_inputs, **generation_params)
# Decode and display response
response = tokenizer.decode(generated_ids[0], skip_special_tokens=True)
print(f"Response:\n{response}")
```
